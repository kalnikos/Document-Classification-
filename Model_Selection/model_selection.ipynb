```python
df1 = path_X_test
df2 = red_y_test
df3 = pd.DataFrame(predictions)
result = pd.concat([df1, df2], axis=1)
result["predictions"] = predictions
result
```

```python
encode = {0:'Admin', 1:'Analyst', 2:'cleaner', 3:'Finance', 4:'hospitality',
       5:'recruitment', 6:'Warehouse', 7:'web developer'}

# Decode
result = result.replace({'Target':encode})
result = result.replace({'predictions':encode})
result.head()
```

### Obtain the highest probability of the predictions for each description

```python
## train set 
import numpy as np
predictions_proba = svm_model.predict_proba(red_X_train).max(axis=1) 
np.sort(predictions_proba, axis=0)
```

```python
## test set
predictions_proba = svm_model.predict_proba(red_X_test).max(axis=1) 
np.sort(predictions_proba, axis=0)
```

### So, I will consider as valid outputs texts with Probability over 0.60. I will use it as a threshold for my application. Observations with a Probability lower than 0.60 will not consider as valid for predictions
